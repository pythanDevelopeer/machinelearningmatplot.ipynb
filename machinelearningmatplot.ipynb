pyplot and object oriented method

```python
heart=pd.read_csv('heart.csv')
heart
```

```python
heart.head()
```

```python
over_40=heart[heart['age']>40]
over_40
```

```python
over_40 = heart[heart['age'] > 40]
over_40.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

object oriented approach

```python
fig,ax=plt.subplots(figsize=(6,5))
over_40.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

life cycle of object 

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter_plot=ax.scatter(x=over_40['age'],
                       y=over_40['chol'],
                       c=over_40['target'])
ax.set(title='Suspectability of Heart Diseaser',
      xlabel='age',
      ylabel='cholesterol')
ax.legend(*scatter_plot.legend_elements(),title='target')
scatter_plot;
```

```python
fig,(plot1,plot2)=plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,20))
scatter_plot=plot1.scatter(x=over_40['age'],
                          y=over_40['chol'],
                          c=over_40['target'],
                          cmap='winter')
plot1.set(title="Heart Disease Problem",
         xlabel='age',
         ylabel='cholestrol')
plot1.legend(*scatter_plot.legend_elements(),title='Survised or Not');
scatter_plot=plot2.scatter(x=over_40['age'],
                          y=over_40['trestbps'],
                          c=over_40['target'],
                          cmap='plasma')
plot2.set(title="Heart Disease Problem",
         xlabel='age',
         ylabel='trestbps')
plot2.legend(*scatter_plot.legend_elements(),title='Survised or Not');
plot2.axhline(y=over_40['trestbps'].mean(),linestyle='--');
plt.style.use('fivethirtyeight')
```
